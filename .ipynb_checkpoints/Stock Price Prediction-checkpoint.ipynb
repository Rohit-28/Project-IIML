```python
#pip install quandl on command line to install quandl
#Importing required Modules
import pandas as pd
import quandl
import datetime
```

```python
#Start Date
start = datetime.datetime(2013,1,1)
end = datetime.date.today()
 
# get Apple, Amazon, Google and Netflix stock data 
#Apple ticker symbol is AAPL
#Amazon ticker symbol is AMZN
#Google ticker symbol is GOOGL
#Netflix ticker symbol is NFLX
#https://www.quandl.com/databases/WIKIP :  Data is colected from here
# First argument is the series we want, second is the source ("GOOGL" for Google! Finance), third is the start date, fourth is the end date
AAPL = "AAPL"
AMZN = "AMZN"
GOOGL = "GOOGL"
NFLX = "NFLX"
apple = quandl.get("WIKI/" + AAPL, start_date=start, end_date=end)
amazon = quandl.get("WIKI/" + AMZN, start_date=start, end_date=end)
google = quandl.get("WIKI/" + GOOGL, start_date=start, end_date=end)
netflix = quandl.get("WIKI/" + NFLX, start_date=start, end_date=end)
 
print(type(apple))
print(type(amazon))
print(type(google))
print(type(netflix))
```

```python
#Print the dimension of Data
print(apple.shape)
print(google.shape)
print(amazon.shape)
print(netflix.shape)
```

```python
print(apple.head())
print(amazon.head())
print(google.head())
print(netflix.head())
```

```python
#For diplaying all columns and rows
from IPython.display import display
pd.options.display.max_columns=None
pd.options.display.max_rows=None
```

```python
frames=[apple,amazon,google,netflix]
```

```python
#Concatinating(adding) all 4 companies data in a single pandas Dataframe
data=pd.concat(frames,keys=['Apple','Amazon','Google','Netflix'])
```
